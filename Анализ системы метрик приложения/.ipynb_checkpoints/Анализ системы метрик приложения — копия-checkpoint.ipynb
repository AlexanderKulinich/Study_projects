К началу лета в приложении появился новый вид контента — спортивный. С его помощью менеджеры стремятся расширить аудиторию, добавив в неё пользователей, которые увлекаются спортом и здоровым образом жизни.

После запуска нового контента и изменения стратегии интерес к продукту в целом вырос, но выручка начала стагнировать. Нужно разобраться, почему это происходит.

## Подготовка данных к когортному анализу

Подготовьте данные к когортному анализу с учётом того, что вы будете изучать окупаемость привлечения пользователей за первые 28 дней.

Для каждого пользователя найдите:
- дату первой сессии (дата привлечения пользователя, `first_dt`) - тут нужна минимальная дата session_start
- канал привлечения (`first_channel`) - channel, соответствующий first_dt;
- стоимость привлечения пользователя (`cac`) - сумма затрат на количество клиентов

Результаты сохраните в новой таблице, например, `profiles`.

Добавьте к данным с событиями и покупками информацию о первом источнике привлечения и дату привлечения пользователя из созданной таблицы. После этого рассчитайте дневной лайфтайм для каждого события или покупки и добавьте в таблицу в виде целого числа. В дальнейшем при проведении когортного анализа используйте информацию о первом источнике привлечения пользователя.

````python
# создаем таблицу с профилями

profiles = (
    visits.sort_values(by="session_start")
    .groupby("user_id")
    .agg(first_dt=("session_start", "min"), first_channel=("channel", "first"))
    .reset_index()
)
profiles["first_dt"] = profiles["first_dt"].dt.normalize()
profiles.info()
````

````python
# считаем САС по каждому устройству в день
cac = (
    costs.groupby(["channel", "dt"])["costs"]
    .sum()
    .reset_index()
    .merge(
        profiles.groupby(["first_channel", "first_dt"])["user_id"].nunique(),
        left_on=["dt", "channel"],
        right_on=["first_dt", "first_channel"],
        how="inner",
    )
    .assign(cac=lambda x: x["costs"] / x["user_id"])
)
cac.columns = ["first_channel", "first_dt", "costs", "user_id", "cac"]
cac.head()
````

````python
# объединяем результаты
profiles = profiles.merge(
    cac[["first_dt", "first_channel", "cac"]],
    on=["first_dt", "first_channel"],
    how="left",
)
````

````python
# добавляем лайфтайм

visits = visits.merge(
    profiles[["user_id", "first_dt", "first_channel"]], on="user_id", how="left"
)
orders = orders.merge(
    profiles[["user_id", "first_dt", "first_channel"]], on="user_id", how="left"
)
````

````python
visits["lifetime"] = (visits["session_start"] - visits["first_dt"]).dt.days.astype(
    "int"
)
orders["lifetime"] = (orders["event_dt"] - orders["first_dt"]).dt.days.astype("int")
visits_28 = visits[visits["lifetime"] < 28]
orders_28 = orders[orders["lifetime"] < 28]
````

Промежуточный вывод: 
- Создали таблицу profiles, где: 
    - посчитали первую дату и первый канал привлечения пользователя 
    - посчитали САС пользователя 
- Также добавили значение лайфтайма к датасетам visits и orders, создали соответствующие им датасеты с лайфтаймом менее 28 дней visits_28 и orders_28.

## Анализ месячной динамики основных метрик продукта

**Задача 1**. Начните анализ данных с изучения динамики активности пользователей и их вовлечённости в продукт. Рассчитайте по всем данным значения DAU, MAU и Stickiness и визуализируйте их. Активными считайте всех пользователей, которые взаимодействовали с приложением.

**Задача 2**. В разрезе каждого месяца привлечения новых пользователей рассчитайте:
- Среднюю стоимость привлечения пользователя (CAC).
- Значение LTV и ROI с учётом покупок, совершённых за 28 дней с момента привлечения.

Здесь и далее при расчёте LTV и ROI учитывайте всех привлечённых пользователей когорты.

Визуализируйте метрики и сформулируйте небольшой промежуточный вывод. Сделайте акцент на общей динамике изменения метрик с учётом появления нового контента в начале лета.

````python
# считаем DAU по всем данным

dau = visits.groupby(visits["session_start"].dt.date)["user_id"].nunique().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(dau["session_start"], dau["user_id"], marker="o")
plt.xlabel("День")
plt.ylabel("DAU")
plt.title("Количество всех активных пользователей")
plt.grid()
plt.show()
````

Промежуточный вывод: отмечаем рост количества активных пользователей

````python
# считаем MAU
# сначала создадим столбец с месяцем, по которому будем агрегировать пользователей

visits["month"] = visits["session_start"].dt.to_period("M").dt.to_timestamp()

mau = visits.groupby("month")["user_id"].nunique().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(mau["month"], mau["user_id"], marker="o")
plt.xlabel("Месяц")
plt.ylabel("MAU")
plt.title("Количество уникальных пользователей по месяцам")
plt.grid()
plt.show()
````

Промежуточный вывод: на графике MAU заметнее, что количество активных пользователей растет после июня, когда были внесены изменения в продукт

````python
# далее считаем Stickiness Rate - для этого возьмем среднее значение DAU за месяц N и разделим его на MAU в месяце N

dau["month"] = pd.to_datetime(dau["session_start"]).dt.to_period("M")

dau_monthly_average = dau.groupby("month")["user_id"].mean().reset_index()
dau_monthly_average["month"] = dau_monthly_average["month"].dt.to_timestamp()
# print(dau_monthly_average.info())
# print(mau.info())
stickiness_table = pd.merge(mau, dau_monthly_average, on="month", how="outer")


stickiness_table["stick"] = 100 * (
    stickiness_table["user_id_y"] / stickiness_table["user_id_x"]
)

plt.figure(figsize=(12, 6))
plt.plot(stickiness_table["month"], stickiness_table["stick"], marker="o")
plt.xlabel("Период")
plt.ylabel("Stickiness Ratio")
plt.title("Показатель Stickiness Ratio по месяцам")
plt.grid()
plt.show()
````

Промежуточный вывод: stickiness вырос с мая по июнь, дальше - рост был с августа с небольшим замдлением в октябре.

````python
# Группируем данные по месяцу, считаем средний САС
cac = profiles.copy()
# Выделяем первый месяц привлечения
cac["first_month"] = cac["first_dt"].dt.to_period("M").dt.to_timestamp()
# Группируем данные по месяцу, считаем средний САС
cac = (
    cac.groupby(["first_month"])
    .agg(
        total_users=("user_id", "nunique"),
        avg_cac=("cac", "mean"),
    )
    .reset_index()
)

profiles.head()
# Оставляем покупки, совершенные в первые 28 дней
ltv = orders_28.copy()
# Выделяем первый месяц привлечения
ltv["first_month"] = ltv["first_dt"].dt.to_period("M").dt.to_timestamp()
# Считаем выручку с пользователей за каждый месяц (по месяцу привлечения)
ltv = (
    ltv.groupby(["first_month"])
    .agg(
        revenue_28=("revenue", "sum"),
    )
    .reset_index()
)
# Добавляем размер когорт и САС
ltv = ltv.merge(cac, on="first_month")
# Считаем LTV на 28 день
ltv["ltv_28"] = ltv["revenue_28"] / ltv["total_users"]
# Рассчитаем ROI за 28 дней:
ltv["roi_28"] = (ltv["ltv_28"] - ltv["avg_cac"]) / ltv["avg_cac"]
````

````python
# сведем все в одну таблицу для расчета метрик

report = all_cohorts.merge(week_costs, on=["cohort_week", "first_channel"], how="left")
report = report.merge(week_revenue, on=["cohort_week", "first_channel"], how="left")

report["costs"] = report["costs"].fillna(0)
report["revenue"] = report["revenue"].fillna(0)

report["cac"] = report["costs"] / report["total_users"]
report["ltv"] = report["revenue"] / report["total_users"]
report["roi"] = (report["ltv"] - report["cac"]) / report["cac"] * 100

report.head()
````

````python
# визуализируем LTV
plt.figure(figsize=(12, 6))
sns.lineplot(data=report, x="cohort_week", y="ltv", marker="o", hue="first_channel")

plt.xlabel("Неделя привлечения")
plt.ylabel("LTV")
plt.title("Динамика LTV по каналам")
plt.axvline(x=pd.Timestamp("2024-06-01"), color="red", linestyle="--")
plt.grid()
plt.show()
````

````python
# визуализируем ROI
plt.figure(figsize=(12, 6))
sns.lineplot(data=report, x="cohort_week", y="roi", marker="o", hue="first_channel")

plt.xlabel("Неделя привлечения")
plt.ylabel("LTV")
plt.title("Динамика ROI по каналам")
plt.axhline(y=0, color="black", linestyle="--", label="Граница окупаемости")
plt.grid()
plt.show()
````

````python
#  подготовим данные для heatmap - сначала соберем количество пользователей в когортах недели, затраты по каналам и САС
cac_data = all_cohorts.merge(week_costs, on=["cohort_week", "first_channel"])
cac_data["CAC"] = cac_data["costs"] / cac_data["total_users"]
# cac_data = cac_data.rename(columns={"user_id": "cohort_size"})
cac_data.head()
````

````python
# считаем LTV через pivot_table
# привязываем заказы к когортам

# profiles["cohort_week"] = profiles["first_dt"].dt.to_period("W").dt.to_timestamp()
# orders_extended = orders.merge(
# profiles[["user_id", "first_dt", "first_channel", "cohort_week"]],
# on="user_id",
# how="left",
# )
# посчитаем лайфтайм в неделях
orders_extended = orders.copy()
orders_extended["cohort_week"] = (
    orders_extended["first_dt"].dt.to_period("W").dt.to_timestamp()
)
orders_extended["lt_weeks"] = ((orders_extended["lifetime"]) / 7).astype(int)


# создаем сводную таблицу
revenue_pivot = orders_extended.pivot_table(
    index=["cohort_week", "first_channel"],
    columns="lt_weeks",
    values="revenue",
    aggfunc="sum",
).fillna(0)

# ограничиваем столбцы - нам нужны только до 10 недели
revenue_pivot = revenue_pivot.loc[:, :10]

# считаем кумулятивную сумму
revenue_cumsum = revenue_pivot.cumsum(axis=1)

print(revenue_cumsum)
````

````python
# напишем функцию для визуализации тепловых карт


def plot_roi_heatmap(channel_name):
    # берем выручку конкретного канала - метод xs выберет строки, где в уровне индекса first_channel стоит значение channel_name
    # в результате получим таблицу выручки только для одного канала, где индекс - это дата когорты
    channel_revenue = revenue_cumsum.xs(channel_name, level="first_channel")

    # берем расходы и размеры когорт этого канала - превращаем столбец с датами когорт в индекс таблицы, чтобы дальше делить одну таблицу на другую
    channel_cac = cac_data[cac_data["first_channel"] == channel_name].set_index(
        "cohort_week"
    )

    # оставляем только общие когорты - то есть оставляем только те недели, где есть полные данные
    common_index = channel_revenue.index.intersection(channel_cac.index)
    channel_revenue = channel_revenue.loc[common_index]
    channel_cac = channel_cac.loc[common_index]

    # считаем LTV (кумулятивная выручка разделить на размер когорты)
    # .div - деление dataframe на серию по индексу
    ltv_table = channel_revenue.div(channel_cac["total_users"], axis=0)

    # считаем ROI: (LTV - CAC) / CAC * 100
    # .sub - вычитание, .div - деление
    roi_table = (
        ltv_table.sub(channel_cac["CAC"], axis=0).div(channel_cac["CAC"], axis=0) * 100
    )

    # Рисуем (оставляем только когорты после апреля для красоты)
    plt.figure(figsize=(15, 10))

    # Красивый формат даты для подписей
    roi_table.index = roi_table.index.strftime("%Y-%m-%d")

    sns.heatmap(
        roi_table, annot=True, fmt=".0f", cmap="RdYlGn", center=0, vmin=-50, vmax=50
    )
    plt.title(f"ROI Heatmap (Cumulative): {channel_name}", fontsize=16)
    plt.xlabel("Неделя жизни", fontsize=12)
    plt.ylabel("Когорта", fontsize=12)
    plt.show()
````

````python
# Запускаем для проблемных каналов
plot_roi_heatmap("FaceBoom")
````

````python
plot_roi_heatmap("TipTop")
````

Промежуточный вывод: 
- Для FaceBoom значения практически не меняются с течением времени, и даже к 10-1 неделе ROI остается сильно отрицательным - этот канал не окупает в принципе. 
- Для TipTop заметна разница между обновлением контента в июне и после. В июне-ноябре ситуация ухудшилась - когорты стартуют с глубокого минуса в 40-60%. Возврат инвестиций увеличился с 1 недели до почти 2 месяцев.
